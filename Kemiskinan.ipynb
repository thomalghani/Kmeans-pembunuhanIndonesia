```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Data Kriminal.csv')
```

```python
dataset
```

```python
#getdata pembunuhan, indeks keparahan kemiskinan, dan indeks kedalaman kemiskinan
dataset_pembunuhan = dataset.iloc[:, [0, 1, -1, -2]]
```

```python
dataset_pembunuhan
```

```python
IPM = dataset.iloc[:, -3]
IPM.values
```

```python
#variabel X untuk kluster pembunuhan berdasarkan indeks keparahan kemiskinan
X = dataset_pembunuhan.iloc[:, 1:3].values
```

```python
X
```

```python
#menentukan jumlah klaster dari data algoritma K-means Pembunuhan 
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter = 300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow method')
plt.xlabel('number of cluster')
plt.ylabel('WCSS')
plt.show()
```

```python
#membuat klaster dengan kmeans dengan jumlah klaster adalah 3
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=100, random_state=0)
y_kmeans = kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
#menampilkan grafik hasil klastering 
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c='red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c='blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c='green', label = 'Cluster 3')
plt.xlabel('Ratio Pembunuhan di daerah')
plt.ylabel('Indeks Keparahan kemiskinan daerah')
plt.legend()
plt.show()
```

# Kesimpulan

Berdasarkan analisa diatas terdapat 3 buah klaster dari data diatas. Ketiga klaster tersebut adalah : 
Cluster 1 (merah) : Memiliki tingkat ratio pembunuhan yang rendah dan indeks keparahan kemiskinan yang rendah
Cluster 2 (Hijau) : Memiliki tingkat ratio pembunuhan yang rendah dan indeks keparahan kemiskinan yang tinggi
Cluster 3 (Biru)  : Memiliki tingkat ratio pembunuhan yang tinggi dan indeks keparahan kemiskinan yang rendah

Berikut akan ditampilkan masing-masing daerah dalam cluster tersebut
